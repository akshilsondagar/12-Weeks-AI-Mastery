```python
# Group By
# “Group similar rows together and perform calculations.”
import pandas as pd

data = {
    "Name": ["Akshil", "Raj", "Priya", "Neha", "Aman"],
    "Department": ["IT", "HR", "IT", "HR", "IT"],
    "Salary": [50000, 40000, 60000, 45000, 55000],
    "Age": [22, 25, 23, 24, 26]
}

df = pd.DataFrame(data)

print(df)
```

```python
df.groupby("Department")["Salary"].mean()
```

```python
# Multiple Aggregations
df.groupby("Department")["Salary"].agg(["mean", "max", "min"])
```

```python
# Count employees in each department
df.groupby("Department")["Name"].count()
```

```python
# Sum
df.groupby("Department")["Salary"].sum()
```

```python
# Multiple Columns Grouping
df.groupby(["Department", "Age"])["Salary"].mean()
```

```python
# Sort Values after GroupBy
df.groupby("Department")["Salary"].mean().sort_values(ascending=False)
```

```python
# GroupBy Multiple Columns
df.groupby("Department")[["Salary", "Age"]].mean()
```

```python
# Custom Function in GroupBy
def salary_range(x):
    return x.max() - x.min()

df.groupby("Department")["Salary"].apply(salary_range)
```

```python
# IMPORTANT AGGREGATION FUNCTIONS
# Function	Meaning
# mean()	Average
# sum()	Total
# count()	Number of rows
# max()	Maximum
# min()	Minimum
# std()	Standard deviation
# median()	Middle value
# nunique()	Unique count
```

```python
# MERGE

# Combine two DataFrames using common columns.
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Akshil", "Raj", "Priya"]
})

print(employees)
```

```python
salary = pd.DataFrame({
    "EmpID": [1, 2, 4],
    "Salary": [50000, 60000, 70000]
})

print(salary)
```

```python
# 1. INNER MERGE 
pd.merge(employees, salary, on="EmpID", how="inner")
```

```python
# LEFT MERGE
pd.merge(employees, salary, on="EmpID", how="left")
```

```python
# RIGHT MERGE
pd.merge(employees, salary, on="EmpID", how="right")
```

```python
# OUTER MERGE
pd.merge(employees, salary, on="EmpID", how="outer")
```

```python
# INNER → matching only

# LEFT  → all left + matching right

# RIGHT → all right + matching left

# OUTER → everything
```

```python
# Merge on Different Column Names
df1 = pd.DataFrame({
    "id": [1, 2],
    "name": ["A", "B"]
})

df2 = pd.DataFrame({
    "emp_id": [1, 2],
    "salary": [50000, 60000]
})

pd.merge(df1, df2,
         left_on="id",
         right_on="emp_id")
```

```python
# Function	Definition
# mean()	Calculates average value
# sum()	Calculates total value
# count()	Counts number of rows/items
# agg()	Applies multiple functions together
# apply()	Applies custom function
# reset_index()	Converts index into normal column
# inner	Keeps only matching rows from both tables
# left	Keeps all rows from left table
# right	Keeps all rows from right table
# outer	Keeps all rows from both tables
# left_on	Specifies left table merge column
# right_on	Specifies right table merge column
# suffixes	Adds suffix to duplicate column names after merge
```
