```python
# Arrays
# An array is a special variable, which can hold more than one value at a time.

cars = ["Ford", "Volvo", "BMW"]
print(cars)
print(cars[0])
# change 0th index value in array
cars[0]="Audi"
print(cars)
```

```python
# length
print(len(cars))
```

```python
# for in loop
for i in cars:
    print(i)
```

```python
# You can use the append() method to add an element to an array.

cars.append("Honda")
print(cars)
```

```python
# You can use the pop() method to remove an element from the array.

cars.pop(1)
print(cars)
```

```python
# You can also use the remove() method to remove an element from the array.

cars.remove("Audi")
print(cars)
```

```python
# 2. Indexing and Slicing (Data Access)

# Indexing: array[row, column]

# Slicing: array[start:stop:step]

data = np.array([[10, 20, 30], 
                 [40, 50, 60], 
                 [70, 80, 90]])

# Access the number 50
print(data[1, 1]) 

# Slice the first two rows and last two columns
# Result: [[20, 30], [50, 60]]
print(data[0:2, 1:3])
```

```python
# Vectorization (Speed)

# Vectorization is the process of performing an operation 
#on an entire array at once rather than looping through 
#individual elements.
prices = np.array([100, 200, 300])

# Instead of: for p in prices: p * 1.1 (Slow)
discounted_prices = prices * 0.9 # Vectorized (Fast)
```

```python
# Vectorization means performing operations on whole arrays WITHOUT loops.
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
```

```python
arr = np.array([1,2,3,4,5,6])

print(arr[arr % 2 == 0])
```

```python
# Broadcasting (Shape Flexibility)
# Broadcasting allows NumPy to work with arrays of different shapes during arithmetic operations.
# Broadcasting means NumPy automatically adjusts array sizes during operations.

grades = np.array([[80, 85], [90, 70]]) # Shape (2, 2)
bonus = np.array([5, 10])               # Shape (2,)

# 'bonus' is broadcasted across both rows
# It becomes [[5, 10], [5, 10]] internally
final_grades = grades + bonus 
# Result: [[85, 95], [95, 80]]
print(final_grades)
```
